1. What is NumPy, and why is it widely used in Python?

```python
import numpy as np
arr = np.array([1, 2, 3, 4])
print(arr * 2)  # Output: [2 4 6 8]
```

2. How does broadcasting work in NumPy?
- Broadcasting is a feature that lets NumPy perform operations on arrays of different shapes without writing loops.

- Example:

```python
import numpy as np
a = np.array([1, 2, 3])
b = 2
print(a + b)  # Output: [3 4 5]
```

Here, even though b is just a number, it is “broadcasted” across all elements in a.

17. Explain the difference between apply() and map() in Pandas.

- map() is used only on Series (mostly for column-wise operations).

- apply() can be used on both Series and DataFrames. It's more flexible and powerful.

- Example:

```python
import pandas as pd

# Example DataFrame
df = pd.DataFrame({
    'Marks': [80, 90, 70, 85]
})

# Using map() to increase each mark by 5
df['Marks'] = df['Marks'].map(lambda x: x + 5)

print(df)

```

```python
# Apply on a Series
df['Marks'] = df['Marks'].apply(lambda x: x * 2)

# Apply on the whole DataFrame to get the max value from each row
df['Max'] = df.apply(lambda row: max(row), axis=1)

print(df)
```

19. How does Pandas simplify time series analysis?

>> Pandas has built-in support for time-series data:

- datetime indexing

- Resampling (resample())

- Time shifting

- Frequency conversion

 Example

```python
import pandas as pd
import numpy as np

# Create a date range
dates = pd.date_range(start='2024-01-01', periods=10, freq='D')

# Create a time series DataFrame
data = pd.DataFrame({
    'Date': dates,
    'Sales': np.random.randint(100, 500, size=(10,))
})

# Set 'Date' as index
data.set_index('Date', inplace=True)

# Resample by week and take sum of sales
weekly_sales = data.resample('W').sum()

print("Original Data:")
print(data)
print("\nWeekly Sales (Resampled):")
print(weekly_sales)
```

20. What is the role of a pivot table in Pandas?

- Pivot tables in Pandas help summarize data, similar to Excel. They allow you to group and aggregate data in a tabular format.

-  Example:

```python
import pandas as pd

# Sample data
data = {
    'Department': ['Sales', 'Sales', 'HR', 'HR', 'IT', 'IT'],
    'Month': ['Jan', 'Feb', 'Jan', 'Feb', 'Jan', 'Feb'],
    'Revenue': [20000, 22000, 15000, 16000, 30000, 31000]
}

df = pd.DataFrame(data)

# Creating a pivot table
pivot_table = pd.pivot_table(df, values='Revenue', index='Department', columns='Month', aggfunc='sum')

print("Pivot Table:")
print(pivot_table)

```

21. Why is NumPy's array slicing faster than Python's list slicing?

- NumPy’s array slicing is faster mainly because NumPy arrays are stored in contiguous blocks of memory, and the operations are performed using optimized C code under the hood. So when we slice an array, NumPy doesn’t actually copy the data—it just creates a view (a reference) of the original array.

- In contrast, Python lists are more flexible but store elements as separate Python objects, which takes more memory and time to access. That’s why slicing lists is slower.

- Example:

```python
 import numpy as np
import time

# NumPy array slicing
arr = np.arange(1000000)
start = time.time()
slice_np = arr[1000:2000]
end = time.time()
print("NumPy slice time:", end - start)
```
